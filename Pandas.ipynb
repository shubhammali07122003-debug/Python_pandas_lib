```python
# import libraries

import pandas as pd
from sqlalchemy import create_engine
```

```python
# load all files

sales = pd.read_excel('sales_data.xlsx')
order = pd.read_csv('orders_data.csv')
emp = pd.read_json('employee_data.json')
```

# Data cleaning

```python
# Remove duplicates

sales.drop_duplicates(inplace=True)
order.drop_duplicates(inplace=True)
emp.drop_duplicates(inplace=True)
```

```python
# handling missing values

sales.fillna(0,inplace=True)
order.fillna(method='ffill',inplace=True)
emp.fillna(emp['salary'].mean(),inplace=True)
```

```python
# Change data types

sales['Order_Date'] = pd.to_datetime(sales['Order_Date'])
sales.info()



order['order_date']= pd.to_datetime(order['order_date'])
order.info()
```

```python
#Feature engineering 

sales.head()

```

```python
sales["Total_Amount"] = (
    sales["Quantity"] *
    sales["Price"] *
    (1 - sales["Discount"] / 100)
)
```

```python
order.head()
```

```python
order["Revenue"] = order["price"] * order["quantity"]
```
